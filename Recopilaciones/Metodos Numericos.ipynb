METODOS NUMERICOS PARA
$f(x)=x-\exp(-x)$


```python
import numpy as np
import matplotlib.pyplot as plt


def fx(x):
  fx = x - (1/np.exp(x))
  return fx
```

```python
#Metodo de Biseccion
cont = 1
a , b = -1 , 2
m=1
while fx(m) != 0:
  fa = fx(a)
  fb = fx(b)
  if fa*fb < 0:
    m = (a + b) / 2
    fm = fx(m)
    if fa*fm < 0:
      a,b = a , m
    elif fb*fm <0:
      a,b = m,b

  cont += 1
print(f"la raiz es {m} ,con {cont} iteraciones")
```

```python
#Metodo Punto Fijo
def g(x):
  gx = 1/np.exp(x)
  return gx

xi = 1
e = 1
cont = 0
while e != 0:
  x1 = g(xi)
  e = abs ((xi - x1) / xi )
  xi = x1
  cont += 1
print(f"la raiz es {xi} , el error es {e} ,con {cont} iteraciones")
```

```python
#Metodo de Newton
def fxx(x):
  fxx = 1 + (1/np.exp(x))
  return fxx

e = 1
xi  = 0.0
cont = 0
while e != 0:
  x1 = xi - ( ( fx(xi) ) / (fxx(xi) ))
  e = abs ((x1 - xi) / x1 )
  xi = x1
  cont += 1
print(f"la raiz es {xi} , el error es {e} ,con {cont} iteraciones")
```

```python
#MEtodo de la Secante

e = 1
xi  = 1
cont = 0
while e != 0:
  x1 = xi - ( ( (xi - (xi -1))*fx(xi) ) / ( fx(xi) - fx((xi-1)) ) )
  e = abs ((x1 - xi) / x1 )
  xi = x1
  cont += 1
print(f"la raiz es {xi} , el error es {e} ,con {cont} iteraciones")
```

```python
# Metodo de la Falsa Posicion
e = 1
a , b = 1, 2
cont = 0
x1 = 0
while e != 0:
  xi = ( (a*fx(b))-(b*(fx(a))) ) / (fx(b) - fx(a))
  if fx(a)*fx(xi) > 0 :
    a = xi
  else:
    b = xi
  e = abs ((x1 - xi) / x1 )
  x1 = xi
  cont += 1

print(f"la raiz es {xi} , el error es {e} ,con {cont} iteraciones")
```

```python
#Metodo de la Secante Modifcada
def fxm(x):
  d = 1/2
  fxm = (  fx(x + (d*x)) - fx(x) ) / (d*x)
  return fxm
e = 1
xi  = 1
cont = 0
while e != 0:
  x1 = xi - ( ( fx(xi) ) / (fxm(xi) ))
  e = abs ((x1 - xi) / x1 )
  xi = x1
  cont += 1
print(f"la raiz es {xi} , el error es {e} ,con {cont} iteraciones")
```
