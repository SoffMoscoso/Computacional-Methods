##Metodo de Lagrange


```python
def lagrange(X, Y, x):
    n = len(X)
    result = 0.0

    for i in range(n):
        term = Y[i]
        for j in range(n):
            if i != j:
                term *= (x - X[j]) / (X[i] - X[j])
        result += term

    return result
```

```python
X = [0, 1, 2]
Y = [1, 3, 2]
```

```python
lagrange(X,Y,1)
```

```python
x = np.arange(0,3)
plt.figure(figsize=(6,5))
plt.plot( X , Y , "o" , color = "red")
plt.plot( x , lagrange( X , Y , x) , "-" , color = "blue" , label = "Interpolacion")
plt.legend()
plt.grid()
plt.show()
```

 Interpolación de Lagrange

El problema de la interpolación polinómica de Lagrange consiste en lo siguiente:
Conocidos los valores de una función $f$ en $n + 1$ puntos distintos $x_i$ , $i =0, 1, · · · , n$ de un intervalo $[a, b]$, nos planteamos obtener un polinomio $P_n$ de grado no superior a $n$, que coincida con la función $f$ en estos $n + 1$ puntos, es decir, la **interpolación polinomial de Lagrange** encuentra un único polinomio que pasa por todos los puntos de datos.
$$P_n(x_i)=f(x_i)$$ para $$i=0,1,...,n$$

Este polinomio se conoce como **polinomio de Lagrange**, $ L (x) $, y como función de interpolación, debe tener la propiedad $ L (x_i) = y_i $ para cada punto del conjunto de datos. Para calcular polinomios de Lagrange, es útil escribirlos como una combinación lineal de **polinomios de base de Lagrange**, $ P_i (x) $, donde
$$
P_i(x) = \prod_{j = 1, j\ne i}^n\frac{x - x_j}{x_i - x_j},
$$

and
$$
L(x) = \sum_{i = 1}^n y_i P_i(x).
$$

Aquí, $ \prod $ significa "el producto de" o "multiplicar".

Notará que por construcción, $ P_i (x) $ tiene la propiedad de que $ P_i (x_i) = 1 $ cuando $ i = j $ y $ P_i (x_j) = 0 $ cuando $ i \ne j $. Como $ L (x) $ es una suma de estos polinomios, puede observar que $ L (x_i) = y_i $ para cada punto, exactamente como se desee.

```python
def Lagrange(X , Y , x):
  Pi = 1
  for i in range(len(X)-1):
    for j in range(len(Y)-1):
      if i == j :
        j += 1
        Pi = (( x - X[j] ) / ( X[i] - X[j] ) )* Pi
      else:
         Pi = (( x - X[j] ) / ( X[i] - X[j] ) )* Pi
  l = 0
  for k in range(len(Y)):
    l = (Y[k] * Pi ) + l
  return l
```

```python
X , Y = [0, 1, 2] ,  [1, 3, 2]
```

```python
x = np.arange(0,3)
plt.figure(figsize=(6,5))
plt.plot( X , Y , "o" , color = "red")
plt.plot( x , Lagrange( X , Y , x) , "-" , color = "blue" , label = "Interpolacion")
plt.legend()
plt.grid()
plt.show()
```

##Metodo Hermite

```python
#Interpolacion de Hermite

def hermite(xi , yi , fi, x):

  def dupli_list(list):
    dupli = []
    for i in range(len(list)):
      dupli.append(list[i])
      dupli.append(list[i])
    return dupli

  xi = dupli_list(xi)
  yi = dupli_list(yi)
  fi = dupli_list(fi)

  divisores = []
  coeficientes =  []
  filtro = yi

  for k in range(1,len(filtro)+1):

    for i in range(len(filtro)-1):

      if k == 1 and i%2 == 0:
        if i == 0:
          y = fi[i]
          divisores.append(y)
        else:
          y = fi[i]
          divisores.append(y)
      else:
        y = ( filtro[i+1] - filtro[i] ) / ( xi[i+k] - xi[i] )
        divisores.append(y)


    coeficientes.append(filtro[0])
    filtro = divisores
    divisores = []

  P = coeficientes[0]
  multi = 1
  for i in range(len(xi)-1):
    multi = (x - xi[i]) * multi
    P = P + ( coeficientes[i+1] * multi)
  return P

#Organizar rango de indices cuando ambos son iguales
```

```python
x , y , fi =[-1,0,1]  , [2,1,2], [-8,0,8]
```

```python
hermite(x,y,fi,1)
```

```python
a = np.arange(-2.5,1.5,0.1)
plt.figure(figsize=(6,4))
plt.plot( x , y , "o", color = "red" , label= "puntos")
plt.plot( a , hermite( x , y ,fi , a ) , "-",  color = "pink" , label= "Interpolacion")
plt.axhline( y= 0 , color = "black")
plt.title("DInterpolacion de hermite")
plt.xlabel("X")
plt.ylabel("fx")
plt.legend()
plt.grid()
plt.show()
```

##Metodo de Newwton


```python
#Metodo de newton
def delta_div(xi , yi , x):
  divisores = []
  coeficientes =  []
  filtro = yi

  for k in range(1,len(filtro)+1):
    for i in range(len(filtro)-1):
      y = ( filtro[i+1] - filtro[i] ) / ( xi[i+k] - xi[i] )
      divisores.append(y)

    coeficientes.append(filtro[0])
    filtro = divisores
    divisores = []

  P = coeficientes[0]
  multi = 1
  for i in range(len(xi)-1):
    multi = (x - xi[i]) * multi
    P = P + ( coeficientes[i+1] * multi)
  return P
```

```python
X = [1,3,4,8]
Y = [2,3,2,10]
```

```python
delta_div(X,Y,1)
```

```python
x = np.arange(12)
plt.figure(figsize=(6,4))
plt.plot( X , Y , "o", color = "red" , label= "puntos")
plt.plot( x , delta_div( X , Y , x ) , "-",  color = "pink" , label= "Interpolacion")
plt.axhline( y= 0 , color = "black")
plt.title("DInterpolacion de Newton")
plt.xlabel("X")
plt.ylabel("fx")
plt.legend()
plt.grid()
plt.show()
```

##Cubix Spline

```python
t = [1,3,4,8]
v = [2,3,2,10]

h = np.diff(t)
delta_y = np.diff(v)

# Calcular segundas diferencias finitas
delta2_y = np.diff(delta_y)

# Calcular las segundas derivadas
dy2_dx2 = np.zeros(len(t))
dy2_dx2[1:-1] = delta2_y / h[1:]


a = v[:-1]
b = delta_y / h - h * dy2_dx2[:-1] / 2
c = dy2_dx2 / 2
d = (dy2_dx2[1:] - dy2_dx2[:-1]) / (6 * h)

# Función para evaluar el spline cúbico en un punto x_val
def cubic_spline(x_val):
    i = np.searchsorted(t, x_val)
    if i == 0:
        i = 1
    elif i == len(t):
        i = len(t) - 1
    xi = t[i - 1]
    yi = a[i - 1]
    bi = b[i - 1]
    ci = c[i - 1]
    di = d[i - 1]
    dx = x_val - xi
    return yi + bi * dx + ci * dx**2 + di * dx**3

# Puntos de interpolación
x_interp = np.linspace(0, 5, 100)
y_interp = [cubic_spline(x_val) for x_val in x_interp]


plt.plot(t, v, 'o', label='Datos Originales')
plt.plot(x_interp, y_interp, label='Interpolación Spline Cúbica')
plt.legend(loc='best')
plt.xlabel('t (s)')
plt.ylabel('v (m/s^2)')
plt.title('Interpolación Spline Cúbica')
plt.grid()
plt.show()
```
